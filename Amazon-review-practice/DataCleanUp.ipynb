```python
# Database
file_path = f"{path}/7817_1.csv"
df = pd.read_csv(file_path)

# Viewing dataset
df
```

```python
import json
from datetime import datetime

# Function to get latest EUR price or fallback to USD
def extract_clean_price(prices_data):
    try:
        prices = json.loads(prices_data) if isinstance(prices_data, str) else prices_data
        # Try EUR first
        eur_prices = [p for p in prices if p.get('currency') == 'EUR']
        if eur_prices:
            eur_prices.sort(key=lambda x: datetime.fromisoformat(x['dateAdded'].replace('Z', '')), reverse=True)
            return eur_prices[0].get('amountMin')

        # Fallback to USD
        usd_prices = [p for p in prices if p.get('currency') == 'USD']
        if usd_prices:
            usd_prices.sort(key=lambda x: datetime.fromisoformat(x['dateAdded'].replace('Z', '')), reverse=True)
            return usd_prices[0].get('amountMin')

        return None
    except Exception as e:
        print(f"Oopsie in row: {e}")
        return None

# Replace the original 'prices' column with the clean value
df['prices'] = df['prices'].apply(extract_clean_price)
```
